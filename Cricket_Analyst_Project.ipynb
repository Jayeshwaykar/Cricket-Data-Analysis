```python
with open("/content/drive/MyDrive/Data Analyst/t20_json.json/t20_wc_batting_summary.json") as f:
  data = json.load(f)

  all_records = []

  for rec in data:
    all_records.extend(rec['battingSummary'])
df_batting = pd.DataFrame(all_records)  
```
